# machine learning

```python
X_train = train[['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
                'humidity', 'windspeed','day', 'hour']]
y_train = train['count']
```

```python
X_test = test[['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
                'humidity', 'windspeed','day', 'hour']]
```

```python
import numpy as np

def rmsle(y_true, y_pred, convertExp=True):
    # 지수변환
    if convertExp:
        y_true = np.exp(y_true)
        y_pred = np.exp(y_pred)

    # 로그변환 후 결측값을 0으로 변환
    log_true = np.nan_to_num(np.log(y_true+1))
    log_pred = np.nan_to_num(np.log(y_pred+1))

    # RMSLE 계산
    output = np.sqrt(np.mean((log_true - log_pred)**2))
    return output
```

```python
from sklearn.linear_model import LinearRegression

linear_reg_model = LinearRegression()
```

```python
log_y = np.log(y_train)  # 타깃값 로그변환
linear_reg_model.fit(X_train, log_y) # 모델 훈련
```

```python
# 변수 중요도 파악
coef = pd.Series(linear_reg_model.coef_, index = X_train.columns)
coef_sort = coef.sort_values(ascending = False)
```

```python
sns.barplot(x=coef_sort.values, y=coef_sort.index)
```

```python
preds = linear_reg_model.predict(X_train)
```

```python
print (f'선형회귀의 RMSLE 값 : {rmsle(log_y, preds, True):.4f}')
```

```python
linearreg_preds = linear_reg_model.predict(X_test) # 테스트 데이터로 예측

submission['count'] = np.exp(linearreg_preds)    # 지수변환(위에서 로그 변환함)
submission.to_csv('result_linear.csv', index=False) # 파일로 저장
```

```python
# XGBoost
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train, log_y)
preds = xgb.predict(X_train)
print (f'XGB| RMSLE & : {rmsle(log_y, preds, True):.4f}')
```

```python
xgb_preds = xgb.predict(X_test)# 테스트 데이터로 예측

submission['count'] = np.exp(xgb_preds)    # 지수변환(위에서 로그 변환함)
submission.to_csv('result_xgb.csv', index=False) # 파일로 저장
```
